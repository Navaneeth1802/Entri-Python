```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('myexcel - myexcel.csv.csv')
df.head()
```

```python
#Getting info about data
df.info()
```

```python
#checking the no of null elements
df.isna().sum()
```

```python
dff=df
```

```python
#QN-1
#Height Function to fix height
import numpy as np
def create_height(height):
    a=np.random.randint(150,180)
    return a
```

```python
dff['hieght']=df['Height'].apply(create_height)
dff.head()
```

```python
#reordering the dataframe for better understanding
dff=dff[['Name','Team','Number','Position','Age','hieght','Weight','College','Salary']]
dff.head()
```

```python
#QN2
#Distribution of Members acorss each team
#Grouping by Team name to get count
Teams=dff.groupby('Team')
Teams['Name'].count().sort_values(ascending=False)
```

```python
#Finding the Percentage Split
Percentage_split=round((Teams['Name'].count()/dff['Name'].count())*100,2)
Percentage_split.sort_values(ascending=False)
```

```python
Percentage_split.plot(kind='bar',figsize=(11,7),color='blue')
plt.xlabel('Percentage of players')
plt.ylabel('Team Names')
```

```python
#QN3
#Segragating them based on position within companies
Player_Segeragation=dff.groupby(['College','Position'])
```

```python
#Finding Top Positions under each team.
Player_Segeragation['Name'].count().sort_values(ascending=False)
```

```python
#QN-4
# Highest Salary Expenditure Team wise
A=Teams['Salary'].sum().sort_values(ascending=False)
x = A.index
y = A.values
plt.figure(figsize=(10, 6))
plt.bar(x, y, color='lightgreen')
plt.title('Highest Salary Expenditure Team-wise')
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

```python
#Salary Per Team and Position
Player_Segeragation['Salary'].sum().sort_values(ascending=False)
```

```python
#QN5
#Correlation Between age and salary
plt.scatter(x='Salary',y='Age',data=dff)
```
